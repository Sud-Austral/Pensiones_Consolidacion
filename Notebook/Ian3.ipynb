```python
import pandas as pd
import os
```

```python
ruta = r"Fondo Tipo y Sexo"
contenido = os.listdir(ruta)
archivos = []
for fichero in contenido:
    if os.path.isfile(os.path.join(ruta, fichero)) and fichero.endswith('.xls'):
        archivos.append(fichero)
archivos
```

```python
diccionario = {}
salida = []
consol = []
n = 0
for t in archivos:
    #print(t)
    df = pd.read_html("Fondo Tipo y Sexo/"+ t)
    if(len(df) == 1):
        df2 = df[0]  
        df2 = df2.rename(columns={'TIPO Y SEXO':'Tipo y Sexo',
                       'FONDO TIPO AMÁS RIESGOSO':'FONDO TIPO A',
                       'FONDO TIPO BRIESGOSO':"FONDO TIPO B",
                       'FONDO TIPO CINTERMEDIO':'FONDO TIPO C',
                       'FONDO TIPO DCONSERVADOR':'FONDO TIPO D',
                       'FONDO TIPO EMÁS CONSERVADOR':'FONDO TIPO E'})
        #print(df2.columns)
    if(len(df) == 3):
        df2 = df[1]  
        df2.columns = df2.iloc[0]
        df2 = df2.drop([0],axis=0)
        df2.rename(columns={ df2.columns[0]: "Tipo y Sexo" }, inplace = True)
        #print(df2.columns)
    
    n = 0
    mes   = t[-6:-4]
    annio = t[-11:-7]
    for i in df2["Tipo y Sexo"]:
        if("AFILIADOS" in i):
            diccionario["afiliacion"]= i
        if("DEPENDIENTES" in i):

            diccionario["dependencia"]=i
        if("INDEPENDIENTES" in i):

            diccionario["dependencia"]=i
        if("Masculino" in i or "Femenino" in i or "S/I" in i):
            diccionario["sexo"]=i
            diccionario["tipo A"]=df2["FONDO TIPO A"][n]
            diccionario["tipo B"]=df2["FONDO TIPO B"][n]
            diccionario["tipo C"]=df2["FONDO TIPO C"][n]
            diccionario["tipo D"]=df2["FONDO TIPO D"][n]
            diccionario["tipo E"]=df2["FONDO TIPO E"][n]
            diccionario["Mes"] = mes
            diccionario["Año"] = annio
            #print(diccionario)
            salida.append(diccionario.copy())
        n = n+1
    dfsalida = pd.DataFrame(salida)
    dfsalida = dfsalida[dfsalida["afiliacion"]!="TOTAL"]
    consol.append(dfsalida.copy())
    #print(df2.columns)
    #break
#df[1]
final = pd.concat(consol)
final.to_excel("Pensiones_Numero_afiliados_segun_Fondo_tipo_sexo.xlsx", index=False)
```
